```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
movies = pd.read_csv('tmdb-movies.csv')
movies.head()
```

### Univariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
plt.figure(figsize=(11,10));
sns.countplot(data=movies, y= "release_year", color = sns.color_palette()[0]);
plt.show()
```

#### insghts : movies highest release in 2014, and the market of movie indusry has been growing since 1960

```python
sns.displot(movies, x="budget", kind="kde")
plt.show()
```

#### insghts : most movies budget are in betweeen 0 le8 and 1 le8

### Bivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
sns.regplot(data=movies, x="budget", y="revenue")
plt.show()
```

#### insghts : positive relation between budget and revenue

```python
plt.figure(figsize=(11,10))
sns.barplot(data=movies, x="release_year", y="revenue",
              color = sns.color_palette()[0])
plt.xticks(rotation=90)
plt.show()
```

#### insghts : revenue has been increasing there is positive Relationship 

### Multivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
sns.scatterplot( data=movies, x="budget", y="revenue", hue="release_year")
plt.show()
```

#### insghts : every new year the budget and revenue of movies increased
